```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

```python
x = np.arange(0, 10, 0.1)
y = x * 2
plt.title("一元一次函数")
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(-10, 10, 0.1)
y = x**2 + 2 * x + 1
plt.title("二元一次函数")
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(-10, 10, 0.1)
y = 2**x
plt.title("指数函数")
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 4, 0.01)
e = math.e
y = e * (x - x + 1)   # **x
plt.title("一字马函数")
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)
plt.title("正弦函数")
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 1, 0.0001)
e = math.e
y = (e**x - 1) / x
plt.title("指数函数导数的常量")
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 1, 0.01)
e = math.e
y = e**(3 * x)
plt.title("e^x的复合函数")
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(-5, 5, 0.00001)
y = (25 - x**2)**0.5
plt.title("圆形")
plt.plot(x, y)
plt.plot(x, -y)
plt.show()
```

```python
x = np.arange(-12, 12, 0.01)
y = (x / np.sin(x))**0.5
plt.title("有趣的隐函数")
plt.plot(x, y)
plt.plot(x, -y)
plt.show()
```

```python
x = np.arange(-8, 2, 0.01)
y = ((2 + x)**3 - 2**3) / x
plt.title("在x=0处间断但极限存在的情形")
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(-6, 6, 0.01)
y = np.sin(np.pi * x) / (x**2 - 1)
y2 = x**2 - 1
y3 = np.sin(np.pi * x)
plt.title("在x=1处间断的情形")
plt.plot(x, y)
plt.plot(x, y2)
plt.plot(x, y3)
plt.ylim(-2, 3)
plt.show()
```

```python
x = np.arange(0, 8, 0.01)
y = x * (8 - x)
plt.title('速度函数')
plt.plot(x, y)
plt.show()
```
