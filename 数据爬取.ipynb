```python
import pandas as pd

names =  ["名称", "性别", "签名", "视频数量","粉丝数",
         "播放数", "点赞数", "总充电人数", "月充电人数",
         "生日", "category1", "category2", "tags"]

df = pd.read_csv("b站up主粉丝量top1万.csv", usecols=[2,3,5]+list(range(9,16))+[22,23,24], header=0, names=names, low_memory=False, converters = {'粉丝数':int})
df.drop_duplicates(inplace=True)
df.sort_values("粉丝数", ascending=False, inplace=True)

df
```

```python
df = df[~df.名称.str.contains('哔哩哔哩')]
df.shape[0]
```

```python
df = df.query("粉丝数>=10000")
df.shape[0]
```

```python
df["粉丝数范围"] = pd.cut(df.粉丝数, bins=[1, 10, 50, 100, 2000], labels=["万粉", "百万前期", "百万后期", "百万粉"], right=False)
```

```python
t = pd.crosstab(df.category1, df.粉丝数范围)
t.style.background_gradient("Reds", axis=0)
```
