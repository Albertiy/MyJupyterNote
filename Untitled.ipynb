```python
class Solution:
    def findMinStep(self, board: str, hand: str) -> int:
        
        turns = 0
        balls = 0 # 不能则-1

        boardSum = {}
        for i in list(board):
            if i in boardSum: boardSum[i]+=1
            else: boardSum[i]=1
        handSum = {}
        for i in list(hand):
            if i in handSum: handSum[i]+=1
            else: handSum[i]=1
        for b in boardSum.keys():
            num = boardSum[b] + (handSum[b] if b in handSum else 0)
            if num < 3: return -1
        for h in handSum:
            if (h in boardSum) or handSum[h] >= 3: handSum[h] = 0

        # 前提条件：手球最多五个，而桌球最多16个。
        # step1: 手球+桌球<3，死局，返回 -1
        # 此时的剩余情况，要么手里无求
        # step2: 
        # 对于有球可入的情况：
        # 1. 绝不会将手球放到无相邻色的位置
        # 2. 间隔死局：指因同色球被另一组同色球分割，而两组球谁先消除都会留下无法消除球的情形，
        # 如：BB=RWRW，HB=RWW，虽然满足step1条件，但无法消除
        # 3. 抉择陷阱：指因非最优消除导致的失败，
        # 如 BB=RRWWRRWWRR，HB=R，只有将R球打入中间的一组RR，才能完成消除；
        # 4. 顺序陷阱：手球中不同色的球，谁先打的问题。
        # 如 BB=RRWWRGGR，HB=WG，若先消除W，则左侧三R连锁消除，右侧R孤立。若先G后W，则顺利消除所有R球。
        # 5. 同色连续球不会超过2个，打球可看做插入，但不在乎插入在前后，因此更像是增加数量，
        # 如 桌球`BB='WWRRBBWW'`，相当于 `[['W',2],['R',2],['B',2],['W',2]]`。
        # 手球没有顺序之分，`HB='GGGGG'`，相当于 `{'G':5}`

        # 尝试过程：
        # 使用回溯法？或分支界限法？界限：所有手球打完；
        # 记录每一步的当前状态，
        # 1. 从手球随机取得一色球，向桌球中随机一位置增加1，复杂度其实不多，
        # 而且后期可以剔除一些情况以跳过死胡同。

        # 驳斥“球必然放入同类中”观点：对于 BWW，手球 BBBW，若每一步依次将B直接放入B处，将导致B最终剩余1个。
        # 因此，小球并不是必然插入同类四周，而是可以放到间隔的区域。只对同类球， 放在附近的效果相同。
        # 驳斥“球不会插入异类球的中间”观点：对于 BGGBBGGB，GGWWW，只有先将W插入中间的BB之间，
        # 然后分别消除G，最后消除W使两组B混合才能完成。
        # 总结：只有当插入同类四周时，才可以合并。不可以对桌面球进行数组合并。
        # 只需让当前手球只插入到同类球的前面，后面直接判断前面是否有同类，有则跳过。
        # 继续思考：每一回合开始时，桌面球不可能有大于2个的连续同类球，因此插入无非三种情况：前面、后面、中间。
        # <del>手球无需转为字典，只需转为列表，能随机和移除即可。</del>

        # 考虑存储每次成功的顺序和位置。

        # bBall = list()  # 废弃：二维列表无法直接浅复制，因此使用一维列表，在字母与数字成对存储。
        # hBall = {} # 废弃：直接构建全排列列表，并合并相同排列，然后循环
        # last = ''
        # for b in board:
        #     if last == b:
        #         bBall[-1] += 1
        #     else:
        #         bBall += [b, 1]
        #         last = b
        bBall = [''] + list(board)   # 正确：只将字符串转换为数组，因其前后都能插值。最好将首项添加一个空值
                            # ，则每个手球只考虑放入后面。
        # print('bBall', bBall)
        # for h in hand:
        #     if h in hBall: hBall[h]+=1
        #     else: hBall[h]=1


        hBall = [x for x in hand if (x in boardSum.keys() or handSum[x]==0)]
        # hBall = [x for x in hand if x in boardSum.keys()] # 错误：不能这么写，因为多余字符也可用于占位
        print('hBall', hBall)

        # print(allSort([1,2,3]))
        temp = allSort(hBall)
        allHand = []
        for t in temp:
            if t not in allHand:
                allHand.append(t)
        # print(allHand)

        print('board: ', bBall)
        res = []
        # TODO
        # 字典在迭代期间不可改变size，不能用pop
        for h in allHand:
            # print('hand: ', h)
            res += allBoard(bBall, h)
            # for hi in h: # 顺序执行
            #     tempBoard = bBall.copy()
            #     for i in range(len(tempBoard)):
            #         if tempBoard[i] == hi: continue # 若前一个元素与待插入元素相同，与插入在此元素之前的情形相同，故略去
            #         tempBoard.insert(i+1, hi)
            #         nextTempBoard = judge(list(tempBoard))
        # print(res)
        return min(res) if res else -1


# 嵌套
def allBoard(board: list, hand: list):
    res = []
    for i in range(len(board)): # 在任意位置插入元素
        tempBoard = board.copy()
        if tempBoard[i] == hand[0]: continue # 前一位置等同插入元素，剪去
        tempBoard.insert(i+1, hand[0])
        nextBoard = judge(list(tempBoard))
        # print(f'{len(hand)}: {hand[0]} => {nextBoard}')
        if nextBoard == ['']:
            res.append(1)
            break # 后续不需要了
        elif len(hand) == 1: pass # res.append(-1)
        else:
            nextHand = hand.copy()
            nextHand.pop(0)
            nextRes = allBoard(nextBoard, nextHand)
            if nextRes:
                res += [n+1 for n in nextRes]
    return res


# 获取列表元素全排列
def allSort(ls: list) -> list:
    if not ls: return ls
    if len(ls) == 1: return [ls]
    res = []
    for idx in range(0, len(ls)):
        temp = ls.copy()
        temp.pop(idx)
        # print(temp)
        res += [ [ls[idx]] + item for item in allSort(temp)]
    return res

# 消除列表中大于等于3的连续项
def judge(ls: list) -> list:
    dirty = False # 脏检查
    before = ''
    length = 1
    for i in range(len(ls)-1, -1, -1):
        if before == ls[i]:
            length += 1
            continue
        elif length >= 3: # 清除之前大于等于3的重复元素
            dirty = True
            for j in range(i+length, i, -1):
                ls.pop(j)
        length = 1
        before  = ls[i]
    if length >= 3:
        dirty = True
        for j in range(i+length-1, -1, -1):
            ls.pop(j)
    # print(ls)
    return judge(ls) if dirty else ls

        
        


# # 遍历手球
# def allHand(hand: dict):
#     for h in hand.keys():
#         hand[h]-=1
#         if hand[h]==0: hand.pop(h)
#         allBoard(board, h): 
    
# # 遍历桌球
# def allBoard(board: list):
#     pass

# # 剪枝
# def pruning():
#     pass
```
